# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1
# passo 1 - Entrar no nosso sistema (entrar no link:https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing )
pyautogui.hotkey('ctrl', 't') # Abre uma nova aba
link = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing' # o link
pyperclip.copy(link) # copiar o link
pyautogui.hotkey('ctrl', 'v') # colar o link na busca 
pyautogui.press('enter') # apertar o Enter

# passo 2 - Navegar no sistema (entrar na pasta exportar)
time.sleep(5) # esperar 5 segundo somente aqui
pyautogui.click(x=344, y=343, clicks=2) # Para clicar onde o mouse estiver posicionado

# passo 3 - baixar o arquivo de vendas
time.sleep(3)
pyautogui.click(x=396, y=361)
pyautogui.click(x=1717, y=192)
pyautogui.click(x=1481, y=624)
time.sleep(5) # para esperar ele fazer o download
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# passo 4 - Calcular faturamento e quantidade de produtos vendidos
import pandas as pd
tabela = pd.read_excel(r'C:\Users\Usuário\Downloads\Vendas - Dez.xlsx')
faturamento = tabela ['Valor Final'].sum()
quantidade = tabela ['Quantidade'].sum()
display(tabela)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# passo 5 - enviar o e-mail para a diretoria
pyautogui.hotkey('ctrl', 't') # abre uma nova aba
link = 'https://mail.google.com/mail/u/0/#inbox'
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

# escrever o nosso e-mail
time.sleep(3)
pyautogui.click(x=38, y=217) #clicar no escrever do e-mail

#escrever para quem to mandando o e-mail
time.sleep(4)
pyautogui.write('nathaliarcruz@gmail.com')
time.sleep(3)
pyautogui.press('tab') # escolher o e-mail
time.sleep(2)
pyautogui.press('tab') # esse é para eu passar para o campo de assunto

#escrever o assunto
assunto = 'Relatório de Vendas'# pq relatório tem ascento.
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab') #para passar pro corpo do e-mail

#escrever o corpo do e-mail
texto = f"""
Prezados , bom dia

o faturamento foi de R$ {faturamento:,.2f}
A quantidade de produtos foi de {quantidade:,}

Abs
Nathalia
"""
pyautogui.write(texto)

#enviar o e-mail
time.sleep(3)
pyautogui.hotkey('ctrl','enter')
```
